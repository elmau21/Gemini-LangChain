- Gemini models are trained to accommodate textual input interleaved with a wide variety of audio and visual inputs, such as natural images, charts, screenshots, PDFs, and videos, and they can produce text and image outputs

## Observations based in the paper:
- The model generates a CoT with multiple samples, and if a consensus above a threshold is reached, the model selects that answer; otherwise, it reverts to a greedy sample.

- The evaluation process involves addressing challenges such as data contamination, and the team performs extensive leaked data analysis to ensure the reported results are scientifically sound. Some benchmarks, such as LAMBADA, are excluded from the report due to identified minor issues. Notably, the HellaSwag benchmark results show that additional finetuning steps on specific website extracts improve the validation accuracy of Gemini Pro and Gemini Ultra, suggesting susceptibility of benchmark results to pretraining dataset composition.

- The team emphasizes the need for more robust and nuanced standardized evaluation benchmarks without leaked data. To address this, Gemini models are evaluated on new held-out datasets, including WMT23 and Math-AMC 2022-2023 problems, as well as internally generated datasets like Natural2Code.

- Despite the challenges in benchmark evaluation, the performance of Gemini models on these benchmarks provides an indication of their capabilities. Gemini Ultra, in particular, demonstrates impressive reasoning and STEM competencies, suggesting potential applications in the educational domain. The model's ability to handle complex mathematical and scientific concepts opens up possibilities for personalized learning and intelligent tutoring systems. The summary encourages further exploration of Gemini's real-world impact on tasks and domains beyond the evaluated benchmarks.

## Chat with Documents

```python
import urllib
import warnings
from pathlib import Path as p
from pprint import pprint

import pandas as pd
from langchain import PromptTemplate
from langchain.chains.question_answering import load_qa_chain
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma

warnings.filterwarnings("ignore")
```

### In Context Information Retrieval




```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
model = ChatGoogleGenerativeAI(model="gemini-pro",temperature=0.3)
```

### Download the document

```python
data_folder = p.cwd() / "data"
p(data_folder).mkdir(parents=True, exist_ok=True)

pdf_url = "https://arxiv.org/pdf/2312.11805.pdf"
pdf_file = str(p(data_folder, pdf_url.split("/")[-1]))

urllib.request.urlretrieve(pdf_url, pdf_file)
```

### Extract text from the PDF

```python
!pip install pypdf
```

```python
pdf_loader = PyPDFLoader(pdf_file)
pages = pdf_loader.load_and_split()
print(pages[3].page_content)
```

```python
context = "\n".join(str(p.page_content) for p in pages[:30])
print("The total words in the context: ", len(context))
```

### Prompt Design - In Context

image.png

```python
prompt_template = """Answer the question as precise as possible using the provided context. If the answer is
                    not contained in the context, say "Please, checkout if the PDF contais the data you want to extract" \n\n
                    Context: \n {context}?\n
                    Question: \n {question} \n
                    Answer:
                  """
prompt = PromptTemplate(
    template=prompt_template,
    input_variables=["context", "question"]
)
```

```python
stuff_chain = load_qa_chain(model, chain_type="stuff", prompt=prompt)
```

### QA Question #1

```python
question = "What is the content about?."

stuff_answer = stuff_chain(
    {"input_documents": pages[7:10], "question": question}, return_only_outputs=True
)
print(stuff_answer)
```

### QA Question #2

```python
question = "Describe the Gemini family."

stuff_answer = stuff_chain(
    {"input_documents": pages[7:10], "question": question}, return_only_outputs=True
)

print(stuff_answer)
```

### RAG Pipeline: Embedding + LLM

image.png

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=0)
context = "\n\n".join(str(p.page_content) for p in pages)
texts = text_splitter.split_text(context)
```

image.png

```python
#Generating the Embeddings
embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")
```

image.png

```python
#Installing chromadb for the vectordb
!pip install chromadb
```

```python
vector_index = Chroma.from_texts(texts, embeddings).as_retriever()
```

```python
question = "Describe the process of Gemini"
docs = vector_index.get_relevant_documents(question)
docs
```

### QA Question #3

```python
stuff_answer = stuff_chain(
    {"input_documents": docs, "question": question}, return_only_outputs=True
)
print(stuff_answer)
```
